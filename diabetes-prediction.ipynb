```python
def mPlotter(r, c, size, _targets, text):
    bg = '#010108'
    palette = ['#df5337', '#d24644', '#f7d340', '#3339FF', '#440a68', '#84206b', '#f1ef75', '#fbbe23', '#400a67']
    font = 'DejaVu Sans'
    
    fig = plt.figure(figsize=size)
    fig.patch.set_facecolor(bg)
    grid = fig.add_gridspec(r, c)
    grid.update(wspace=0.5, hspace=0.25)
    
    __empty_diff = ((r * c) - 1) - len(_targets)
    axes = []
    
    for i in range(r):
        for j in range(c):
            ax = fig.add_subplot(grid[i, j])
            ax.set_facecolor(bg)
            axes.append(ax)
    
    for idx, ax in enumerate(axes):
        ax.spines['left'].set_visible(False)
        ax.spines['top'].set_visible(False)
        ax.spines['right'].set_visible(False)
        
        if idx == 0:
            ax.spines["bottom"].set_visible(False)
            ax.tick_params(left=False, bottom=False)
            ax.set_xticklabels([])
            ax.set_yticklabels([])
            ax.text(0.5, 0.5, text, horizontalalignment='center', verticalalignment='center',
                    fontsize=18, fontweight='bold', fontfamily=font, color="#fff")
        elif idx - 1 < len(_targets):
            # Ensure x-tick labels (names of bars) are shown
            ax.tick_params(axis='x', colors='#fff', rotation=45)  # Rotate for better visibility if needed
            ax.tick_params(axis='y', colors='#fff')
            ax.spines["bottom"].set_visible(True)
            ax.set_title(_targets[idx - 1].capitalize(), fontsize=14, fontweight='bold', fontfamily=font, color="#fff")
            ax.grid(color='#fff', linestyle=':', axis='y', zorder=0,  dashes=(1,5))
        else:
            ax.set_visible(False)  # Hide unused axes
    
    def cb(ax):
        # Placeholder callback function to modify axes properties if necessary
        ax.set_xlabel("")
        ax.set_ylabel("")
    
    return axes, palette, cb
```

# Model Development

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.metrics import accuracy_score, classification_report
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from sklearn.model_selection import train_test_split
```

```python
df.drop(['PERSON NAME', 'so/wo', 'Date of Birth', 'REGISTRATION_QULIFICATION', 'BloodPressureStatus', 'Weight (20-200Kg)', 'HeightInInch (31-78)', 'Body Mass Index', 'Hip (9.8-59Inches)', 'BloodSugarReportStatus', 'Hip Waist Status'], axis=1, inplace=True)
```

```python
# Define the target column and feature columns
target_column = 'Abdominal Status'
X = df.drop(columns=[target_column])
y = df[target_column]
```

```python
# Preprocessing pipeline for numerical features
numerical_pipeline = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),  # Impute missing values
    ('scaler', StandardScaler())  # Standardize features
])

# Preprocessing pipeline for categorical features (already label encoded)
categorical_pipeline = Pipeline(steps=[
    ('onehot', OneHotEncoder(handle_unknown='ignore'))  # One-hot encode
])
```

```python
# Combine numerical and categorical pipelines
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_pipeline, X.select_dtypes(include=['float64', 'int64']).columns.tolist()),
        ('cat', categorical_pipeline, X.select_dtypes(include=['object', 'category']).columns.tolist())
    ])
```

```python
# Define the models to evaluate
models = {
    'AdaBoost': AdaBoostClassifier(n_estimators=100, learning_rate=0.001),
    'Support Vector Classifier': SVC(probability=True),
    'K-Nearest Neighbors': KNeighborsClassifier(),
    'Naive Bayes': GaussianNB()
}
```

```python
from sklearn.model_selection import cross_val_score, cross_val_predict
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import pandas as pd

# Initialize a dictionary to store results
results = {}

# Loop through the models and perform evaluations with different CV splits
for model_name, model in models.items():
    # Create the pipeline with the preprocessor and the model
    model_pipeline = Pipeline(steps=[
        ('preprocessor', preprocessor),
        ('classifier', model)
    ])
    
    # Fit the model pipeline on the entire dataset
    model_pipeline.fit(X, y)
    
    # Predict on the training data to get training accuracy
    y_pred = model_pipeline.predict(X)
    training_accuracy = accuracy_score(y, y_pred)
    training_precision = precision_score(y, y_pred, average='binary')
    training_recall = recall_score(y, y_pred, average='binary')
    training_f1 = f1_score(y, y_pred, average='binary')
    
    # Initialize the results for this model
    results[model_name] = {
        'Training Accuracy': training_accuracy,
        'Training Precision': training_precision,
        'Training Recall': training_recall,
        'Training F1': training_f1
    }
    
    # Perform cross-validation with different k values
    for cv in [2, 5, 10]:
        y_pred_cv = cross_val_predict(model_pipeline, X, y, cv=cv)
        results[model_name][f'k{cv} Accuracy'] = accuracy_score(y, y_pred_cv)
        results[model_name][f'k{cv} Precision'] = precision_score(y, y_pred_cv, average='binary')
        results[model_name][f'k{cv} Recall'] = recall_score(y, y_pred_cv, average='binary')
        results[model_name][f'k{cv} F1'] = f1_score(y, y_pred_cv, average='binary')

# Convert the results dictionary to a DataFrame for easy manipulation and visualization
results_df = pd.DataFrame(results).T.reset_index()
results_df.rename(columns={'index': 'Model'}, inplace=True)
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import roc_curve, auc
from sklearn.model_selection import StratifiedKFold
from sklearn.pipeline import Pipeline
from sklearn.ensemble import AdaBoostClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB

# Define classifiers with basic settings
classifiers = {
    'AdaBoost': AdaBoostClassifier(n_estimators=100, learning_rate=0.001),
    'Support Vector Classifier': SVC(probability=True),
    'K-Nearest Neighbors': KNeighborsClassifier(),
    'Naive Bayes': GaussianNB()
}

# Stratified K-Fold configuration
kfolds = [2, 5, 10]

# Prepare the plot
fig, axs = plt.subplots(1, len(kfolds), figsize=(18, 6))
colors = ['b', 'g', 'r', 'c', 'm']  # Different colors for each model

# Iterate through k-fold configurations
for i, k in enumerate(kfolds):
    cv = StratifiedKFold(n_splits=k, shuffle=True, random_state=42)
    mean_fpr = np.linspace(0, 1, 100)  # Common FP rates for averaging TP rates

    for (label, classifier), color in zip(classifiers.items(), colors):
        tprs = []  # True positive rates
        aucs = []  # Area under curve scores
        for train_idx, test_idx in cv.split(X, y):
            X_train, X_test = X.iloc[train_idx], X.iloc[test_idx]
            y_train, y_test = y.iloc[train_idx], y.iloc[test_idx]

            # Assuming the use of a preprocessor
            model = Pipeline([
                ('preprocessor', preprocessor),  # Use your actual preprocessor
                ('classifier', classifier)
            ])
            model.fit(X_train, y_train)
            y_proba = model.predict_proba(X_test)[:, 1]
            fpr, tpr, _ = roc_curve(y_test, y_proba)
            tprs.append(np.interp(mean_fpr, fpr, tpr))
            tprs[-1][0] = 0.0  # Ensure start at zero
            roc_auc = auc(fpr, tpr)
            aucs.append(roc_auc)

        mean_tpr = np.mean(tprs, axis=0)
        mean_tpr[-1] = 1.0  # Ensure end at one
        mean_auc = auc(mean_fpr, mean_tpr)
        std_auc = np.std(aucs)
        axs[i].plot(mean_fpr, mean_tpr, color=color, label=f'{label} (AUC = {mean_auc:.2f} Â± {std_auc:.2f})', lw=2, alpha=0.8)

    axs[i].plot([0, 1], [0, 1], linestyle='--', lw=2, color='gray')
    axs[i].set_xlim([0, 1])
    axs[i].set_ylim([0, 1.05])
    axs[i].set_xlabel('False Positive Rate')
    axs[i].set_ylabel('True Positive Rate')
    axs[i].set_title(f'ROC Curves for {k}-Fold Cross-Validation')
    axs[i].legend(loc="lower right")

plt.tight_layout()
plt.show()
```

# Senstivity Comparison

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import recall_score
from sklearn.model_selection import StratifiedKFold
from sklearn.pipeline import Pipeline
from sklearn.ensemble import AdaBoostClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB

# Define classifiers with basic settings
classifiers = {
    'AdaBoost': AdaBoostClassifier(n_estimators=100, learning_rate=0.001),
    'Support Vector Classifier': SVC(probability=True),
    'K-Nearest Neighbors': KNeighborsClassifier(),
    'Naive Bayes': GaussianNB()
}

# Stratified K-Fold configuration
kfolds = [2, 5, 10]

# Prepare the plot
fig, axs = plt.subplots(1, len(kfolds), figsize=(18, 6))
colors = ['b', 'g', 'r', 'c']  # Different colors for each model

# Iterate through k-fold configurations
for i, k in enumerate(kfolds):
    cv = StratifiedKFold(n_splits=k, shuffle=True, random_state=42)
    for (label, classifier), color in zip(classifiers.items(), colors):
        sensitivities = []
        for train_idx, test_idx in cv.split(X, y):
            X_train, X_test = X.iloc[train_idx], X.iloc[test_idx]
            y_train, y_test = y.iloc[train_idx], y.iloc[test_idx]

            # Assuming the use of a preprocessor
            model = Pipeline([
                ('preprocessor', preprocessor),  # Use your actual preprocessor
                ('classifier', classifier)
            ])
            model.fit(X_train, y_train)
            y_pred = model.predict(X_test)
            sensitivity = recall_score(y_test, y_pred)
            sensitivities.append(sensitivity)

        mean_sensitivity = np.mean(sensitivities)
        std_sensitivity = np.std(sensitivities)
        axs[i].bar(label, mean_sensitivity, yerr=std_sensitivity, color=color, alpha=0.6, capsize=5)

    axs[i].set_ylim([0, 1])
    axs[i].set_xlabel('Model')
    axs[i].set_ylabel('Sensitivity')
    axs[i].set_title(f'Sensitivity for {k}-Fold Cross-Validation')

plt.tight_layout()
plt.show()
```

**Let's Save SVC and make predictions**

```python
from sklearn.model_selection import KFold
from sklearn.pipeline import Pipeline
from sklearn.svm import SVC
from joblib import dump
from sklearn.metrics import accuracy_score

# Create a KFold object for 2 splits
kf = KFold(n_splits=2, shuffle=True, random_state=42)

# Initialize variables to store the best model and highest accuracy found across folds
best_model = None
highest_accuracy = 0

# Manually iterate over the two folds
for train_index, test_index in kf.split(X):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]

    # Create the SVC pipeline
    svc_pipeline = Pipeline([
        ('preprocessor', preprocessor),
        ('classifier', SVC(probability=True))
    ])
    
    # Train the SVC model
    svc_pipeline.fit(X_train, y_train)

    # Evaluate the model on the test fold
    y_pred = svc_pipeline.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    
    # Save the model if it has the best accuracy so far
    if accuracy > highest_accuracy:
        best_model = svc_pipeline
        highest_accuracy = accuracy

# Save the best model found on the k=2 splits
dump(best_model, 'svc_model_k2.joblib')
```

```python
# Loading the trained model
from joblib import load
svc_model = load('svc_model_k2.joblib')

# Preparing the new unseen data (same features and preprocessing)
new_data = pd.DataFrame({
    'Age': [55],
    'REGISTRATION_GENDER': ['Male'],
    'REGISTRATION_MARITAL_STS': ['Married'],
    'BSR (50-550mg/dL)': [140],
    'BP_S (50-250)': [130],
    'BP_D (30-200)': [80],
    'Waist (9.8-59Inches)': [50],
    'Waist-Hip Ratio': [0.84],
    'Body Mass Index Status': ['Normal'],
    'PEFR Status': ['low']
})

# Making predictions
predictions = svc_model.predict(new_data)
probabilities = svc_model.predict_proba(new_data)

# Printing predictions with confidence scores
for i, prediction in enumerate(predictions):
    print(f'Prediction: {prediction}, Confidence Score: {probabilities[i][prediction]:.2f}')
```
